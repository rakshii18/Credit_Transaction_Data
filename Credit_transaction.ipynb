### RF Classification with hyperparameter tuning

```python
#Performing RF Classification and tuning parameters.
from sklearn.model_selection import GridSearchCV, cross_val_score
from sklearn.ensemble import RandomForestClassifier
rfc=RandomForestClassifier()
```

```python
#Selecting parameters for hyper tuning.
n_estimators=[2,5,10,25,50]
max_depth=[5,10,15,20,25,30]
criterion=['entropy']
```

```python
rfc2=RandomForestClassifier(n_estimators=gs.best_estimator_.get_params()['n_estimators'],criterion='entropy',max_depth=gs.best_estimator_.get_params()['max_depth'])
```

### Partial Dependece Plots

```python
from sklearn.inspection import plot_partial_dependence
```

```python
plot_partial_dependence(gbm3,z_train2,features=[17, 14],target=1)
```

```python
plot_partial_dependence(rfc2,z_train2,features=[17, 14],target=1)
```

The partial dependence plots of the 2 methods is very interesting. The PDP of that of GBM is a straight line while that of RF
is not. This means that features in GBM is displaying linear dependence, while that in RF is not dependent on the target.

Here, since we are working on the imbalanced dataset, the accuracy scores will not actually depict the performance of the model. The model for gbm does not really reflect if it works incase of any frauds occur (based on the f-1 score).
